```python
import pandas as pd
from urllib.request import urlopen

def extractLink(url):
    try: 
        return urlopen(url).geturl()
    except:
        return ""

extract_df = pd.read_csv("grouped_links.csv", sep=",", header=0)

extract_df.columns = ['link', 'occurences']

extract_df["new_links"] = extract_df["link"].apply(extractLink)

extract_df
```

```python
extract_df.to_csv("extracted.csv", header=0)
```
